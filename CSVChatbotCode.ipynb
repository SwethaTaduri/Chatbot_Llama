```python
import os
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = 'VMware/open-llama-7b-v2-open-instruct'

tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast = False)

model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype= torch.float16, device_map="auto", offload_folder="offload")

prompt_template = "You are an honest, courteous, and helpful helper. Always respond in a way that is both safe and helpful.  Nothing hurtful, immoral, sexist, racist, poisonous, deadly, or unlawful should be included in your responses. Please make sure that your comments are constructive and socially neutral. Give an explanation for a question that defies logic or lacks factual coherence rather than providing an incorrect response. Please refrain from spreading incorrect information if you are unsure about the answer to a question. Write a response that appropriately completes the request.\n\n### Instruction:\n{instruction}\n\n### Response:"

def LLM_chatbot(prompt):
  inputt = prompt_template.format(instruction=prompt)
  input_ids = tokenizer(inputt, return_tensors="pt").input_ids.to("cuda")

  output1 = model.generate(input_ids, max_length=512)
  input_length = input_ids.shape[1]
  output1 = output1[:, input_length:]
  output= tokenizer.decode(output1[0])

  return output
```

```python
def withOutIndex(prompt):
  print("\nRetrieving With Out Index response now...\n")
  return LLM_chatbot(prompt)
```

```python
print(withOutIndex("What are the different types of breast cancer?"))
```

```python
import os

import sys

from pinecone import Pinecone

from langchain.document_loaders.csv_loader import CSVLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Pinecone
from langchain.llms import Replicate
from langchain.chains import ConversationalRetrievalChain
```

```python
os.environ['REPLICATE_API_TOKEN'] = 
os.environ["PINECONE_API_KEY"] = 
```

```python
loader = CSVLoader(file_path="/content/TCGA_Reports.csv", encoding='utf-8')
docs = loader.load()
```

```python
divider = CharacterTextSplitter(chunk_size=2000, chunk_overlap=0)
chunks = divider.split_documents(docs)
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
from langchain_pinecone import PineconeVectorStore
store = PineconeVectorStore.from_documents(chunks, embeddings, index_name="reports")
```

```python
language_model = Replicate(
    model="a16z-infra/llama13b-v2-chat:df7690f1994d94e96ad9d568eac121aecf50684a0b0963b25a41cc40061269e5",
    input={"temperature": 0.65, "max_length": 4000}
)
```

```python
chain_pipeline = ConversationalRetrievalChain.from_llm(
    language_model,
    store.as_retriever(search_kwargs={'k': 2}),
    return_source_documents=True
)
```

```python
previous_context = []
```

```python
def withIndex(prompt):
  print("\nRetrieving With Index response now...\n")
  llm_response = chain_pipeline({'question': prompt, 'chat_history': previous_context})
  previous_context.append((prompt, llm_response['answer']))
  return ('Answer: ' + llm_response['answer'] + '\n')
```

```python
print(withIndex("Can you give details about one of the tumors from the TCGadataset?"))
```

```python
query = "Which patient has the largest tumor?"
docs = store.similarity_search(query)
print(docs)
```
